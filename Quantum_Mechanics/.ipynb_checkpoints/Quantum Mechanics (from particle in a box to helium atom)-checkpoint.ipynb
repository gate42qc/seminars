# <center>Wave Equation</center>

### The analytical solution to the wave equation that describes a 1-D string can obtained ''separation of variables'' technique. The solution is given in general form by:

## <center>$A(x,t)=\sum_n c_n \cos(\frac{n\pi ct}{l})\sin(\frac{n\pi x}{l})$</center>


### $\sin(\frac{n\pi x}{l})$ is the time independent portion 
### $\cos(\frac{n\pi ct}{l})$ is the time dependent portion. 
### $n$ is an integer that comes from satisfying the boundary conditions. 
### $\rightarrow$ the wave can be expressed as a sum of "Eigenstates" or "Harmonics", each one characterized by a different integer number $n$

## Let's see what the time-independent harmonics look like for $l=10$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define your function
def harmonic_time_independent(x,n,l):
    return np.sin(n*np.pi*x/l)
# create variables
l = 10 
x = np.linspace(0,l,1000)
y = np.zeros((3,len(x)))
# use a for loop for the multiple harmonics
plt.figure(figsize=(12,6))
for n in range(4):
    # plot stuff
    y[n-1] = harmonic_time_independent(x,n,l)
    plt.plot(y[n-1])
plt.show()
```

## And the time-dependent harmonics ($l=10$, $c=0.1$)

```python
# define the function
def harmonic_time_dependent(t,c,n,l):
    return np.cos(n*np.pi*c*t/l)
# create variables
l = 10.0
c = 0.1
t = np.linspace(0,100,1000)
y = np.zeros((4, len(t)))
# use a for loop for the multiple harmonics
plt.figure(figsize=(12,6))
for n in range(1, 5):
    # plot stuff
    y[n-1] = harmonic_time_dependent(t,c,n,l)
    plt.plot(y[n-1])
    plt.legend(y[n-1])
plt.show()
```

# <center>Superposition and Interference </center>

### Consider the following wave, built from two similar waves:
## <center>$A(x,t)=\cos(\frac{\pi ct}{l})\sin(\frac{\pi x}{l})+\cos(\frac{2\pi ct}{l})\sin(\frac{2\pi x}{l})$</center>

```python
# define your function
def wave_solution(x,t,c,n,l):
    solution = harmonic_time_dependent(t,c,n,l)*harmonic_time_independent(x,n,l)
    return solution
# initialize variables
t = np.arange(0, 100, 0.1)
l = 10
c = 0.5
x = l/3
# sum both waves
wave = wave_solution(x,t,c,1,l) + wave_solution(x,t,c,2,l)
# plot
plt.figure(figsize=(12,6))
plt.plot(t, wave)
plt.show()
```

## We can de-construct any function into simple waves, but also reconstruct it via the amplitude coefficients.

```python
import scipy
import scipy.integrate as integrate
```

```python
def square_function(x, l):
    if x <= l/3.0 or x >= l*2.0/3:
        y = 0
    else:
        y = 1
    return y

def projection_integrand(x,n,l):
    integrand = np.sqrt(2.0/l)*np.sin((n*np.pi*x)/l)*square_function(x,l)
    return integrand

# initial variables
l=10.0
x = np.arange(0,l,0.01)
y = np.array([square_function(xi, l) for xi in x])
square_approx = np.zeros(len(x))
# for loop over n
for n in range(0,40):
    # project amplitudes, sum components
    c_n,_ = integrate.quad(projection_integrand, 0,l,args=(n,l))
    square_approx = square_approx + c_n * np.sqrt(2.0/l)*np.sin((n*np.pi*x)/l)
# plot
plt.figure(figsize=(12,6))
plt.plot(x,y)
plt.plot(x,square_approx)
plt.title('Square wave')
plt.show()
```

# <center>QM postulates</center>

## 1. A wavefunction $\psi$ for a given system contains all of the information about the entire quantum system
## 2. The wavefunction is interpreted as a probability amplitude, with the absolute square of the wavefunction, $\psi^*(r,t)\psi(r,t)$, interpreted as the probability density at time t
## 3. For every observable property $A$ of a system there is a quantum mechanical operator $\hat{A}$
## 4. When using a normalized wave function $\psi$, the average value of an observable is given by its expectation value
## <center>$\langle \psi | \hat{A} | \psi \rangle = \int \psi^{\ast}\hat{A}\psi dx$</center>
## 5. The wavefunction of a system evolves in time according to the time-dependent Schr√∂dinger equation
## <center>$i\hbar\frac{\partial\psi}{\partial t}= H\psi$</center>

## <center>$H = p^2/2m + V$ </center>
## <center>$\psi_{pib}(x, l, n) = \sqrt{\frac{2}{l}} sin( \frac{n\pi x}{l})$</center>
## <center>$E_{pib}(n,l, m)=\frac{n^2 \hbar^2 \pi^2}{2 m l^2}$</center>

## some helper functions to simulate quantum systems (particle in a box)

```python
import numpy as np
# first three functions
def pib_eigenfunction(x, l, n):
    eigenfunction = np.sqrt(2/l)*np.sin(n*np.pi*x/l)
    return eigenfunction

def prob_density(psi_x):
    density = np.conjugate(psi_x)*psi_x
    return density
    
def pib_energy(n, l, hbar=1, m=1):
    energy = (n**2 * hbar**2 * np.pi**2)/(2 * m * l**2)
    return energy
    
# finite diff function
def finite_diff(y,dx):
    # create variables
    derivative = []
    # edge case i=0
    edge_derivative_0 = (y[1] - y[0])/dx
    derivative.append(edge_derivative_0)
    # iterate over 'middle points"
    for i in range(1,len(y)-1):
        mid_derivative = (y[i+1] - y[i-1])/(2*dx)
        derivative.append(mid_derivative)
    # edge case i=n-1
    edge_derivative_end = (y[-1] - y[-2])/dx
    derivative.append(edge_derivative_end)
    
    return derivative
```

```python
from scipy.integrate import simps
# create your normalize wfn function
def normalize_wfn(x, psi_x):
    norm_sq = simps(np.conjugate(psi_x) * psi_x, x=x)
    return psi_x / np.sqrt(norm_sq)
# initialize variables
L = 10.0
x = np.arange(0,L,0.01)
psi_x = 0
# do a for loop to sum all 4 eigenfunctions
for n in range(1, 5):
    psi_x += qw.pib_eigenfunction(x, L, n)

# get pdf, normalize for psi_x
pdf = qw.prob_density(psi_x)

# calculate norms before and after
psi_normed = normalize_wfn(x, psi_x)
norm_before = simps(np.conjugate(psi_x) * psi_x, x=x)
norm_after = simps(np.conjugate(psi_normed) * psi_normed, x=x)
print('Norm before:',norm_before)
print('Norm after:',norm_after)
# plot
plt.figure(figsize=(12,6))
plt.plot(x, psi_x)
plt.plot(x, pdf)
plt.show()
```

# <center>Time-Dependent Schrodinger Equation: setup</center>

## Consider the wavefunction as a superposition of the first two eigenstates of the pib:
## <center>$\psi_0(x)=c_1\psi_1(x)+c_2\psi_2(x)$</center>
## with coefficients $c_1=c_2=1/\sqrt(2)$

```python
# cnt evolve funtion
def cnt_evolve(cn_0, t, E_n, hbar = 1.0):
    return cn_0 * np.exp(-1j*E_n*t/hbar )
#initialize variables
L = 10
x = np.arange(0,L,0.01)
# construct first state
psi1_x = qw.pib_eigenfunction(x,L,1)
c1 = 1/np.sqrt(2)
E1 = qw.pib_energy(1,L)
# construct second state
psi2_x = qw.pib_eigenfunction(x,L,2) 
c2 = 1/np.sqrt(2)
E2 = qw.pib_energy(2,L)
# superposition
psi0 = c1*psi1_x + c2*psi2_x
# norm
psi0_norm = qw.wfn_norm(x,psi0)
print('Norm is ',psi0_norm)
#plot
plt.figure(figsize=(12,6))
plt.plot(x,psi0)
plt.show()
```

```python
#initialize variables
L = 10
x = np.arange(0,L,0.01)
t = np.linspace(0,50,100)
psi = np.zeros((len(x), len(t)))
# construct first state
c1_0 = 1/np.sqrt(2)
psi1_x = qw.pib_eigenfunction(x,L,1)
E1 = qw.pib_energy(1, L)
# construct second state
c2_0 = 1/np.sqrt(2)
psi2_x = qw.pib_eigenfunction(x,L,2)
E2 = qw.pib_energy(2, L)
# iterate over time
for i in range(len(t)):
    # time evolve coefficients
    c1_t = qw.cnt_evolve(c1_0, t[i], E1, hbar=1)
    c2_t = qw.cnt_evolve(c2_0, t[i], E2, hbar=1)
    # save snapshot of psi
    psi[:,i] = c1_t*psi1_x + c2_t*psi2_x

# time-plot
plt.figure(figsize=(12,6))
qw.time_plot(x,psi,t,t_step=2)
plt.show()
```

# <center>Observable operators: position and momentum</center>

## <center>$\hat{x} =x$</center>
## <center>$\hat{p} = -i \hbar \frac{d}{dx}$</center>
## <center>$\langle \psi | \hat{A} | \psi \rangle = \int \psi^{\ast}\hat{A}\psi dx$</center>

```python
# define momentum operator function
def momentum_operator(psi_x, dx, hbar = 1):
    return -1j/hbar * qw.finite_diff(psi_x, dx)

# initialize variables
L = 10
dx = 0.01
x = np.arange(0,L,dx)
c1 = 1.0/np.sqrt(2)
c2 = 1.0/np.sqrt(2)
psi1_x = qw.pib_eigenfunction(x,L,1)
psi2_x = qw.pib_eigenfunction(x,L,2)
psi0 = c1*psi1_x + c2*psi2_x

# Expectation of position operator
x_integrand = np.conjugate(psi0)*x*psi0
exp_position = qw.complex_simps(x_integrand, x)
print('Expectation of position is', exp_position)

# Expectation of momentum operator
p_integrand = np.conjugate(psi0) * momentum_operator(psi0, dx, hbar = 1)
exp_momentum = qw.complex_simps(p_integrand, x)
print('Expectation of momentum is', exp_momentum)
```

# <center>1-D Time-Independent Schr√∂dinger Equation (TISE)</center>

## <center>$\hat{H}\psi(\hat{r}) = E \psi(\hat{r})$</center>
## 1. The Hamiltonian HÃÇ  is the operator that encodes the physics for particles in quantum systems (at least for the kinetic and the potential part).
## 2. The equation is an eigenvalue problem ($Av=Œªv$), eigenvalues and eigenvectors are 'natural' solutions to matrices. In QM, these solutions are going to be energy values and eigenstates of a quantum system.
## <center>$\hat{H}=\hat{T}+\hat{V}$</center>
## <center>$\hat{T} = \frac{\hat{p}^2}{2m}=-\frac{\hbar^2}{2m}\frac{\partial^2 }{\partial x^2}.$</center>
## <center>$\frac{\partial^2 y_i}{\partial x^2} = \begin{cases}
            \frac{y_{i+1} - 2 y_{i} + y_{i-1}}{ {\Delta x}^2} \text{, if  } 0 < i < n -1 \\
            \frac{y_{i+1} - 2 y_{i}}{ {\Delta x}^2} \text{, if } i=0\\
            \frac{ - 2 y_{i} + y_{i+1}}{ {\Delta x}^2} \text{, if } i=n-1
        \end{cases}$</center>
        
## <center>$\hat{T}=\begin{pmatrix}
-2t & t & 0 & 0 \\
t & -2t & t & 0 \\
0 & t & -2t & t \\
0 & 0 & t & -2t
\end{pmatrix},\quad\hat{V}=\begin{pmatrix}
v(x_0) & 0 & 0 & 0 \\
0 & v(x_1) & 0 & 0 \\
0 & 0 & v(x_2) & 0 \\
0 & 0 & 0 & v(x_3)
\end{pmatrix}$</center>

```python
def linear_ramp(x):
    b = 2
    return b * x
v_x = linear_ramp(x)
x = np.arange(0,L,0.01)
plt.figure(figsize=(12,6))
plt.plot(x, v_x)
plt.show()
```

```python
# initial variables
L=10.0
dx = x[1] - x[0]
# define kinetic operator function
def kinetic_operator(x, m=1, h_bar=1):
    # precalculate variables like dx, t and intialize T
    dx = x[1]-x[0]
    t = -(h_bar**2) / (2.0 * m*dx**2)
    T = np.zeros((len(x), len(x)))
    # iterate over x
    for i in range(len(x)):
        # consider first diagonal elements
        T[i][i] = -2 * t
        # then side diagonal elements, consider edge cases (i=0) or (i=n-1)
        if i == 0:
            T[i][i + 1] = t
        elif i == len(x) - 1:
            T[i][i - 1] = t
        else:
            T[i][i + 1] = t
            T[i][i - 1] = t
    return T
# get T, then V and then H
T = kinetic_operator(x)
V = np.diag(v_x)
```

## <center>The 2-D Harmonic potential</center>
## <center>$V(x,y) = \frac{1}{2}k_x(x - x_0)^2 + \frac{1}{2}k_y(y - y_0)^2$</center>

```python
# function definition
def harmonic_potential_2D(xx, yy, kx, ky, x0=0, y0=0):
    pot = 1/2*kx*(xx-x0)**2 + 1/2*ky*(yy-y0)**2
    return pot
# intialize variables
kx = 0.05
ky = 0.05
x = np.linspace(-4, 4, 50)
y = np.linspace(-4, 4, 50)
# meshgrid and evaluate the function
xx, yy = np.meshgrid(x,y)
vxy = harmonic_potential_2D(xx, yy, kx, ky, x0=0, y0=0)
# 3d plotting
qw.plot_3d_surface(xx,yy,vxy)
plt.show()
# contouring
plt.clf()
plt.figure(figsize=(12,6))
qw.plot_contours(xx,yy,vxy)
plt.show()
```

# <center>Infrared Spectroscopy</center>

## <center><font color='red'>A spectrum is how molecules say hello to you!!!</font></center>
## <center>New concepts</center>
## A time-dependent Hamiltonian.
## Solutions to the harmonic oscillator interacting with a time-dependent field.
## Transition probabilities.
## The infrared spectrum.

## <center>$\hat{H}(t) = -\frac{\hbar^2}{2m}\frac{d}{dx^2}+\frac{1}{2}m\omega_0^2x^2+\frac{2\pi}{\lambda}E_0\sin(\omega_f\;t)x$</center>
## Solution (perturbation theory)
## <center>$\psi(x,t) = \psi_0(x) + \frac{iE_0\frac{2\pi}{\lambda}}{2\sqrt{2m\hbar\omega_0}}\left[\frac{e^{-i(\omega_0 - \omega_f)t}-1}{\omega_0 - \omega_f}+\frac{e^{i(\omega_0 + \omega_f)t}-1}{\omega_0 + \omega_f}\right]\;\psi_1(x)$</center>

```python
# function code
def time_dependent_psi(x, t, omega_f, omega_0 = 1, lam = 1, E_0 = 1, m=1, hbar=1):
    psi_0 = qw.harmonic_oscillator_wfn(x,0)
    term1 = 1j*E_0*(2*np.pi/lam) / (2*np.sqrt(2*m*hbar*omega_0))
    term2 = (np.exp(-1j*(omega_0-omega_f)*t)-1) / (omega_0-omega_f) + (np.exp(1j*(omega_0+omega_f)*t)-1) / (omega_0+omega_f)
    psi_1 = qw.harmonic_oscillator_wfn(x, 1)
    psi_x_t = psi_0 + term1 * term2*psi_1
    return psi_x_t
```

## <center>Wavepacket dynamics</center>

```python
#initialize variables
x = np.arange(-3, 3+0.01, 0.01)
t = np.linspace(0,4,250)
omega_f=3.0

psit = np.zeros((len(x),len(t)))
# iterate over time
for indt,ti in enumerate(t):
    psit[:,indt] =  qw.time_dependent_psi(x, ti, omega_f)
# time plot
plt.figure(figsize=(12,6))
qw.time_plot(x,psit,t,8)
plt.show()
```

## <center>Excited states overlap</center>

## <center>   $c_1(t) =  \int_{-\infty}^{\infty}\psi^*_1(x)\;\psi(x,t)\;dx$</center>
## <center>$=\frac{iE_0\frac{2\pi}{\lambda}}{2\sqrt{2m\hbar\omega_0}} \left[ \frac{e^{-i(\omega_0 - \omega_f)t}-1}{\omega_0 - \omega_f}+\frac{e^{i(\omega_0 + \omega_f)t}-1}{\omega_0 + \omega_f} \right]$</center>

```python

# function definition
def excited_overlap(t, omega_f, omega_0 = 1, lam = 1, E_0 = 1.0, m=1, hbar=1):
    energy_term= 1j*E_0*( (2*np.pi)/lam)
    denom_term = 2*np.sqrt(2*hbar*omega_0)
    term1 = (np.exp(-1j*(omega_0-omega_f)*t)-1) / (omega_0-omega_f)
    term2 = (np.exp(1j*(omega_0+omega_f)*t)-1 ) / (omega_0+omega_f)
    value = (energy_term/denom_term) *(term1+term2)
    return value
# initialize variables
t = np.arange(0,10,0.01)
omegas = [3.0,5.0,8.0]
prob_excited_state = np.zeros((len(omegas),len(t)))
# iterate over omegas
for indx, omega_f in enumerate(omegas):
    c1_t = excited_overlap(t, omega_f)
    prob_excited_state[indx,:] = np.abs(c1_t**2)
    plt.figure(figsize=(12,6))
    plt.plot(t, prob_excited_state[indx,:],label='omega={}'.format(omega_f))
    # plot
    plt.xlabel('$t$')
    plt.legend()
    plt.ylabel('$|C_1(t)|^2$')
    plt.show()
```

## <center>Calculating Infrared Spectra</center>
## <center>$A_{0 \to 1}=\int_{0}^{t}c_1(t)\;dt$</center>

### Now, let's create an array of many values of $\omega_f$, and for each one of these values of $\omega_f$ we'll calculate $A_{0 \to 1}$, as shown above. By squaring $A_{0 \to 1}$, we'll get the value of the intensity at that particular $\omega_f$, which corresponds to the average transition probability, $P_{0 \to 1}$:

## <center>$P_{0 \to 1}=|A_{0 \to 1}|^2$</center>


```python
# initalize variables
omegas = np.linspace(0,3,1195)
t = np.arange(0,30,0.01)
#exercise
ir_spectrum = np.zeros_like(omegas)
# iterate over omegas
for indx, omega_f in enumerate(omegas):
    # calculate overlaps, integrate, save result
    c1_t = qw.excited_overlap(t, omega_f)
    transition_amplitude = qw.complex_simps(c1_t,t)
    ir_spectrum[indx] = np.abs(transition_amplitude)**2
#plot
plt.figure(figsize=(12,6))
plt.plot(omegas, ir_spectrum)
plt.xlabel('$\omega_f$')
plt.ylabel('Absorption intensity')
plt.title('IR Spectrum')
plt.show()
```

### <font color='red'>The intensity of the transition is proportional to the average transition probability</font>   <font color='blue'>and the plot of $P_{0 \to 1}$ as a function of $\omega_f$ constitutes a spectrum</font>
